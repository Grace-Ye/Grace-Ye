```python
import numpy as np
import pandas as pd
from pandas import DataFrame,Series 
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
import os 
print(os.getcwd())
```

```python
data_2015=pd.read_csv('2015.csv')
data_2015
data_2015=data_2015.drop(['Region','Standard Error','Dystopia Residual'],axis=1)
data_2015
```

```python
data_2016=pd.read_csv('2016.csv')
data_2016
data_2016=data_2016.drop(['Region','Lower Confidence Interval','Upper Confidence Interval','Dystopia Residual'],axis=1)
data_2016
```

```python
data_2017=pd.read_csv('2017.csv')
data_2017
data_2017=data_2017.drop(['Whisker.high','Whisker.low','Dystopia.Residual'],axis=1)
data_2017=data_2017.rename(columns={'Happiness.Rank':'Happiness Rank','Happiness.Score':'Happiness Score','Economy..GDP.per.Capita.':'Economy (GDP per Capita)','Health..Life.Expectancy.':'Health (Life Expectancy)','Trust..Government.Corruption.':'Trust (Government Corruption)'})
data_2017
```

```python
data_2018=pd.read_csv('2018.csv')
data_2018
data_2018=data_2018.rename(columns={'Country or region':'Country','Social support':'Family','Overall rank':'Happiness Rank','Score':'Happiness Score','GDP per capita':'Economy (GDP per Capita)','Healthy life expectancy':'Healthy (life expectancy)','Freedom to make life choices':'Freedom','Perceptions of corruption':'Trust (Government Corruption)'})
data_2018
```

```python
data_2019=pd.read_csv('2019.csv')
data_2019
data_2019=data_2019.rename(columns={'Overall rank':'Happiness Rank','Country or region':'Country','Score':'Happiness Score','GDP per capita':'Economy (GDP per Capita)','Social support':'Family','Healthy life expectancy':'Healthy (life expectancy)','Freedom to make life choices':'Freedom','Perceptions of corruption':'Trust (Government Corruption)'})
data_2019
```

```python
dataframe=[data_2015,data_2016,data_2017,data_2018,data_2019]
dataframe=pd.concat(dataframe)
dataframe['Health (Life Expectancy)']=dataframe['Health (Life Expectancy)'].fillna(0)
dataframe['Healthy (life expectancy)']=dataframe['Healthy (life expectancy)'].fillna(0)
dataframe['Life Expectancy']=dataframe['Health (Life Expectancy)']+dataframe['Healthy (life expectancy)']
dataframe=dataframe.drop(['Health (Life Expectancy)','Healthy (life expectancy)'],axis=1)
dataframe
```

```python
dataframe.describe()
```

```python
dataframe=dataframe.drop(['Country'],axis=1)
for column in dataframe.columns:
    plt.figure()
    sns.distplot(dataframe[column])
```

```python
for column in dataframe.columns:
    plt.figure()
    sns.boxplot(dataframe[column],showmeans=True)
```

```python
sns.lmplot(x='Trust (Government Corruption)',y='Happiness Rank',data=dataframe)
```

```python
dataframe['log(Trust (Government Corruption))']=np.log(dataframe['Trust (Government Corruption)'])
dataframe['log(Happiness Score)']=np.log(dataframe['Happiness Score'])
dataframe
sns.lmplot(x='log(Trust (Government Corruption))',y='log(Happiness Score)',data=dataframe)
```

```python
Q1 = dataframe['Trust (Government Corruption)'].quantile(0.25)
Q3 = dataframe['Trust (Government Corruption)'].quantile(0.75)
IQR = Q3 - Q1
print(IQR)
df_filter = (dataframe['Trust (Government Corruption)'] >= Q1 - 1.5*IQR) & (dataframe['Trust (Government Corruption)'] <= Q3 + 1.5*IQR)
df_filter=dataframe.loc[df_filter]
df_filter
```

```python
sns.lmplot(x='log(Trust (Government Corruption))',y='log(Happiness Score)',data=df_filter)
```

```python
corrMatrix=dataframe.corr()
corrMatrix
sns.heatmap(corrMatrix,annot=True)
plt.show()
```

```python
Y=dataframe['Happiness Score']
dataframe_feature=['Happiness Score','Economy (GDP per Capita)','Family','Freedom','Life Expectancy']
X=dataframe[dataframe_feature]
X.describe()
```

```python
from sklearn.tree import DecisionTreeRegressor
happiness_model=DecisionTreeRegressor(random_state=1)
happiness_model.fit(X, Y)
X.head()
happiness_model.predict(X.head())
```

```python
X.head()
```

```python
from sklearn.metrics import mean_absolute_error
predict_happiness_score=happiness_model.predict(X)
mean_absolute_error(Y,predict_happiness_score)
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_Y, test_Y = train_test_split(X, Y, test_size=0.5,random_state = 0)
happiness_model = DecisionTreeRegressor()
happiness_model.fit(train_X, train_Y)
happiness_pred=happiness_model.predict(test_X)
mean_absolute_error(test_Y,happiness_pred)
```

```python
plt.scatter(test_Y,happiness_pred)
```
